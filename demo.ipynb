```python
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder
import random
import pickle
```

```python
s="B.Com student in a university Sukhdev. I am learning Business Management. I have done interships in the past. I know about finance and bank management and knows trading stuff also. Looking for a job in bank or finance related job. I aim to consult people and manage the stocks and finance related things."
```

```python
model = keras.models.load_model('chat_model')

# load tokenizer object
with open('tokenizer.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)

# load label encoder object
with open('label_encoder.pickle', 'rb') as enc:
    lbl_encoder = pickle.load(enc)

max_len = 4000
result = model.predict(keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([s]),
                                                                  truncating='post', maxlen=max_len))
```

```python
result=result.reshape((result.shape[1],1))
```

```python
result.shape
```

```python

maxi=-1
for i in result:
    if i>maxi:
        maxi=i
```

```python
tag = lbl_encoder.inverse_transform([np.argmax(result)])
```

```python
tag=tag[0]
```

```python
tag
```

```python
df
```

```python
jobs=[]
for i in df.index:
    if df['Title'][i]==tag:
        jobs.append(df.loc[i])
        break
```
